```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
choose_state = soup.find("select", id="parentState").find_all("option")
all_states = choose_state[1:58]
```

```python
# defining button_2 (to click on select state)
button2 = browser.find_by_id('stateSelection').find_by_id('parentState')
# defining button_3 (to click on apply search)
button4 = browser.find_by_id('applySearch')
# defining button_list to make table in list view
button_list = browser.find_by_id('dataView').find_by_id('vList')

# extracting all abbreviations for states. this will enable use to create a loop for clicking.
for item in all_states:
    print(item["value"])
    button3 = browser.find_by_id('stateSelection').find_by_id('parentState').find_by_value(item["value"])
    time.sleep(1)
    button3.click()
    time.sleep(1)
    button4.click()
    time.sleep(1)
    button_list.click()
    time.sleep(1)
    
```
